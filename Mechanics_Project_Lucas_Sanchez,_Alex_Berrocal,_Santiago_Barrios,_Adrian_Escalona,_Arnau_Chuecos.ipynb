```python
import matplotlib.pyplot as pyplot
import numpy as np
import math

# First of all we define our coordinates system, that will remain constant during the project and will be very useful to keep our goal in mind
# Earth's radial direction = z, we consider that Earth's surface is z = 0
# Earth's latitude = y
# Earth's longitude = x

G = 6.674*(10**-11); M = 5.972*(10**24); EarthRadius = 6.371*(10 ** 6); W = 2*math.pi/(24*3600)

def getLastValue(array):
    i = len(array) - 1
    while i > 0:
        if array[i] != 0:
            return array[i]
        i -= 1
    return -1

def addLeadingZeros(number):
  if number < 10:
    return "0" + str(number)
  else:
    return str(number)

def printCoordinates(lat, lon, latReference, longReference):
    lat = np.abs(lat)
    lon = np.abs(lon)

    latMinutes = (lat - int(lat)) * 60
    latSeconds = (latMinutes - int(latMinutes)) * 60

    lonMinutes = (lon - int(lon)) * 60
    lonSeconds = (lonMinutes - int(lonMinutes)) * 60

    if lat - int(lat) > 0:
        lat = int(lat)

    if latMinutes - int(latMinutes) > 0:
        latMinutes = int(latMinutes)

    if latSeconds - int(latSeconds) > 0:
        latSeconds = np.round(latSeconds, 1)

    if lon - int(lon) > 0:
        lon = int(lon)

    if lonMinutes - int(lonMinutes) > 0:
        lonMinutes = int(lonMinutes)

    if latSeconds - int(latSeconds) > 0:
        latSeconds = np.round(latSeconds, 1)

    latMinutes = addLeadingZeros(latMinutes)
    latSeconds = addLeadingZeros(latSeconds)
    lonMinutes = addLeadingZeros(lonMinutes)
    lonSeconds = addLeadingZeros(lonSeconds)

    print("Coordinates: " + str(lat) + " " + latMinutes + "' " + latSeconds + "\" " + str(latReference) + " " + str(lon) + " " + lonMinutes + "' " + latSeconds + "\" " + str(longReference) + "." )

# FASE DE SUBIDA
# TABLE 1:
def get_u(g, isp):
    return g*isp

def get_g(r):
    return (G*M)/((r+EarthRadius)**2)

def increase_velocity(u, mo, mf, g, t):
    return u*math.log(mo/mf) - g*t

def increase_height(deltav, ho):
    return ((G*M) / ((G*M/(EarthRadius+ho)) - deltav**2 / 2) - (EarthRadius+ho))

# Stage 1:
Mt1 = 23230  # kg
Mp1 = 15960  # kg
T1 = 899  # kN
tbo1 = 61  # s
hbo1 = 34  # km
Isp1 = 267
g1 = get_g(0)

# Stage 2:
Mt2 = 7270  # kg
Mp2 = 5260  # kg
Isp2 = 284  # s
T2 = 266  # kN
tbo2 = 65  # s
hbo2 = 96  # km
g2 = get_g(hbo1*1000)

# Stage 3:
Mt3 = 2010  # kg
Mp3 = 1660  # kg
Isp3 = 275  # s
T3 = 76  # kN
tbo3 = 59  # s
hbo3 = 600  # km
g3 = get_g(hbo2*1000)

# 2
# For each of the three stages we must aproximate:
# the value of u(exhaust gas velocity);
# the increment of velocity of the missile after each stage;
# assuming that: vo1 = 0, vo2=vf1, vo3=vf2, obtain all final velocities and heights.
# Burn-out rate = k, Ff = 0

# 2.a
u1 = get_u(g1, Isp1)
u2 = get_u(g2, Isp2)
u3 = get_u(g3, Isp3)

print("Exercice 2a\n")
print("Gas exhaust velocity during stages:")
print("-> First stage: ", np.round(u1), "m/s")
print("-> Second stage: ", np.round(u2), "m/s")
print("-> Third stage: ", np.round(u3), "m/s")
print("__________________________________________________________________________\n")

# 2.b
deltav1 = increase_velocity(u1, Mt1, Mt1 - Mp1, g1, tbo1)
deltav2 = increase_velocity(u2, Mt2, Mt2 - Mp2, g2, tbo2)
deltav3 = increase_velocity(u3, Mt3, Mt3 - Mp3, g3, tbo3)
print("Exercice 2b\n")
print("Increment of velocity during stages:")
print("-> First stage: ", np.round(deltav1), "m/s")
print("-> Second stage: ", np.round(deltav2), "m/s")
print("-> Third stage: ", np.round(deltav3), "m/s")
print("__________________________________________________________________________\n")

# 2.c
vo1 = 0
vo2 = deltav1  # First Stage = vf1
vo3 = vo2 + deltav2  # Second Stage = vf2
finalVelocity = vo3 + deltav3  # Third Stage = vf3

print("Exercice 2c\n")
print("Final velocity at stage:")
print("-> First stage: ", np.round(vo2), "m/s")
print("-> Second stage: ", np.round(vo3), "m/s")
print("-> Third stage: ", np.round(finalVelocity), "m/s")
print("__________________________________________________________________________\n")
print("Final height at stage:")
print("-> First stage: ", hbo1, "km")
print("-> Second stage: ", hbo2, "km")
print("-> Third stage: ", hbo3, "km")
```

Once the final velocity after stage 3 is reached a small engine is used to rotate the missile, so that

Use Euler's method to determine its range, its deviation due to the Coriolis effect and the approximate point on the Earth's surface in which Minuteman II will impact.
You can assume that air friction and the centrifugal acceleration are negligible.

```python
import matplotlib.pyplot as plt
import numpy as np

G = 6.674*(10**-11); M = 5.972*(10**24); EarthRadius = 6.371*(10 ** 6); W = 2*math.pi/(24*3600)

# 3
# We launch Minuteman II from Malmstrom AFB, Montana
# After stage 3 we use a small engine to rotate (E-W, 25 angle)
# Using Eulers method to determine its range,
# deviation due to Coriolis effect,
# and an aproximate impact site
# Centrifugal Force and Air Friction are negligible

ho = 600 * 1000
hf = 0

finalVelocity = 9381 # Data from previous exercise
g3 = 9.530161919936615 # Data from previous exercise

voy =  finalVelocity * np.cos(25 * (2 * np.pi / 360))
voz = finalVelocity * np.sin(25 * (2 * np.pi / 360))

t1 = (voy + np.sqrt(voy ** 2 - 4 * g3 * (-ho))) / g3
t2 = (voy - np.sqrt(voy ** 2 - 4 * g3 * (-ho))) / g3
# 2 times because of the 2nd grade ecuation
if t1 > t2:  # We choose the greatest time in order to avoid negative numbers
    t = t1
else:
    t = t2

tf = t
dt = round(tf / 2000, 1) # We set the differential of Time

time = np.arange(0, tf, dt) # Set of the time array
timeLen = len(time)
# We set all the arrays from the length of the time array
X = np.zeros(timeLen); Y = np.zeros(timeLen); Z = np.zeros(timeLen)
Vx = np.zeros(timeLen); Vy = np.zeros(timeLen); Vz = np.zeros(timeLen)

Vmod = np.zeros(timeLen)

latitud = np.zeros(timeLen)
longitud = np.zeros(timeLen)
latdeg = np.zeros(timeLen)
longdeg = np.zeros(timeLen)
# Set of initial conditions
X[0] = 0; Y[0] = 0; Z[0] = ho
Vx[0] = 0; Vy[0] = - voy; Vz[0] = voz

latdeg[0] = 47.50
longdeg[0] = -111.2

latitud[0] = latdeg[0] * np.pi / 180
longitud[0] = longdeg[0] * np.pi / 180
# Euler loop
i = 1
while i < timeLen and Z[i-1] >= hf:
  # We calculate in each iteration the accelerations
    ax = (2 * W * np.sin(latitud[i - 1]) * Vy[i-1])
    ay = -(2 * W * (np.cos(latitud[i - 1]) * Vz[i - 1] +  2 * W * np.sin(latitud[i - 1]) * Vx[i - 1]))
    az = (2 * W * np.cos(latitud[i - 1]) * Vy[i - 1] - get_g(Z[i - 1]))
  # With accelerations we get the velocity in each axis
    Vx[i] = Vx[i - 1] + ax * dt
    Vy[i] = Vy[i - 1] + ay * dt
    Vz[i] = Vz[i - 1] + az * dt

    Vmod[i] = np.sqrt(Vx[i] ** 2 + Vy[i] ** 2 + Vz[i] ** 2) # Module of velocity

    X[i] = X[i - 1] + Vx[i - 1] * dt
    Y[i] = Y[i - 1] + Vy[i - 1] * dt
    Z[i] = Z[i - 1] + Vz[i - 1] * dt

    dlatitud = - (X[i] - X[i - 1]) / EarthRadius
    latitud[i] = latitud[i - 1] + dlatitud

    dlong = (Y[i] - Y[i - 1]) / (EarthRadius * np.cos(latitud[i - 1]))
    longitud[i] = longitud[i - 1] + dlong

    # We transform the angles to fit the correct intervals
    while longitud[i] > np.pi:
      longitud[i] -= 2 * np.pi

    while longitud[i] < -np.pi:
      longitud[i] += 2 * np.pi

    while latitud[i] > np.pi / 2:
      latitud[i] -= np.pi

    while latitud[i] < -np.pi / 2:
      latitud[i] += np.pi

    longdeg[i] = longitud[i] * 180 / np.pi
    latdeg[i] = latitud[i] * 180 / np.pi
    i += 1

maxRange = np.round(np.abs(getLastValue(Y)) / 1000, 3)
coriolisDeviation = np.round(np.abs(getLastValue(X)) / 1000, 3)
# We get the last non-null value of the arrays
finalX = getLastValue(longdeg)
finalY = getLastValue(latdeg)
finalZ = 0

latReference = "N"
longReference = "E"

if finalX < 0:
    longReference = "W"
if finalY < 0:
    latReference = "S"

print("Approximate impact point (decimal): ", np.round(finalY, 6), np.round(finalX, 6))
printCoordinates(np.abs(finalY), np.abs(finalX), latReference, longReference)
print("Coriolis deviation: ", coriolisDeviation, "km")
print("Range: ", maxRange, "km")

lastTime = time[i - 1]

plt.figure(1)
plt.plot(time, X)
plt.xlim(0,lastTime)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title("X vs. Time")

plt.figure(2)
plt.plot(time, Y)
plt.xlim(0,lastTime)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title("Y vs. Time")

plt.figure(3)
plt.plot(time, Z)
plt.xlim(0,lastTime)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title("Z vs. Time")

plt.figure(4)
plt.plot(time, longdeg)
plt.xlim(0, lastTime)
plt.xlabel('Time (s)')
plt.ylabel('Longitude ()')
plt.title("Longitude vs. Time")

plt.figure(5)
plt.plot(time, latdeg)
plt.xlim(0, lastTime-1)
plt.ylim(45,60)
plt.xlabel('Time (s)')
plt.ylabel('Latitude ()')
plt.title("Latitude vs. Time")

plt.show()

#kml file to graph our route in google earth
K=open("trayectoria.kml","w")
K.write('''<kml xmlns="http://earth.google.com/kml/2.0">
<Document>
<Style id="normalPlaceMarker">
<IconStyle>
<Icon>
<href>root://icons/palette-3.png</href>
<x>96</x>
<y>160</y>
<w>32</w>
<h>32</h>
</Icon>
</IconStyle>
</Style>
<Placemark>
<description> </description>
<name> </name>
<styleUrl>#yellowLineGreenPoly </styleUrl>
<LineString>
<extrude>1</extrude>
<tessellate>1</tessellate>
<altitudeMode>absolute</altitudeMode>
<coordinates>''')

i=0
while i < timeLen and Z[i] == 0:
    K.write(str(latitud[i]))
    K.write("," + str(longitud[i]))
    K.write("," + str(Z[i]))
    i += 1

K.write(''' </coordinates>
</LineString>
</Placemark>
</Document>
</kml>''')

print() # Used to avoid Collaborative problems and errors
```
